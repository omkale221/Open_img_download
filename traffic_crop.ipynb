```python
from pandas import read_csv
from os import listdir
import cv2
```

```python
def read_all_images(data_dir, annot_dir):
    df = read_csv(annot_dir)
    data_folders = listdir(data_dir)
    print(data_folders)
    img_list = df.ImageID.tolist()                                
    for folder in data_folders:
        if folder.endswith('.csv') or folder.endswith('.ipynb'):
            continue
        image_folder_dir = data_dir + '/' + folder
        images = listdir(image_folder_dir)
        print(images)
        for image in images:
            if image.endswith('.jpg') and image[:-4] in img_list:
                image_path = image_folder_dir + '/' + image
                image_ = cv2.imread(image_path)
                height, width, channels = image_.shape
                print(width,height)
                image_data = df.loc[(df['ImageID'] == image[0:-4])]
                counter = 0
                for index, row in image_data.iterrows():
                    print(image)
                    xmin, xmax, ymin, ymax = unconvert(width, height, row['XMin'], row['XMax'], row['YMin'], row['YMax'])
                    print(xmin, xmax, ymin, ymax)
                    cropped = image_[ymin:ymax, xmin:xmax]
                    filename = 'E:/CEMTREX/traffic_cropped/' + image[:-4] + str(counter) + '.jpg'
                    cv2.imwrite(filename, cropped)
                    counter += 1
```

```python
## converts the normalized positions  into integer positions
def unconvert(width, height, x, y, w, h):
#     xmax = int((x*width) + (w * width)/2.0)
#     xmin = int((x*width) - (w * width)/2.0)
#     ymax = int((y*height) + (h * height)/2.0)
#     ymin = int((y*height) - (h * height)/2.0)
    xmin = int(x*width)
    xmax = int(y*width)
    ymin = int(w*height)
    ymax = int(h*height)
    return (abs(xmin), abs(xmax), abs(ymin), abs(ymax))
```

```python
data_dir = 'E:/CEMTREX/vehicle_val'
annot_dir = 'O:/daaaCEMTREX/openImgCSv/v6/labels/openImages_traffic/train_traffic_annotations.csv'
read_all_images(data_dir, annot_dir)
```
